```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
file_path = './casos_with_embedding.csv'
```

```python
work_df = pd.read_pickle(file_path)
```

```python
work_df
```

```python
embeddings = work_df['embedding'].to_list()
```

```python
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(embeddings)

work_df['cluster'] = kmeans.labels_
```
